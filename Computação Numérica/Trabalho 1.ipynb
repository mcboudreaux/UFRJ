#Questão 1


```python
def calcular_operacoes(x, ex, y, ey):
    # Calcular as operações
    soma = x + y
    subtracao = x - y
    produto = x * y
    divisao = x / y

    # Calcular os erros absolutos
    erro_abs_soma = ex + ey
    erro_abs_subtracao = ex + ey
    erro_abs_produto = abs(y * ex) + abs(x * ey)
    erro_abs_divisao = (ey * abs(x)) / (y ** 2)

    # Calcular os erros relativos
    erro_rel_soma = erro_abs_soma / abs(soma)
    erro_rel_subtracao = erro_abs_subtracao / abs(subtracao)
    erro_rel_produto = erro_abs_produto / abs(produto)
    erro_rel_divisao = erro_abs_divisao / abs(divisao)

    return soma, erro_abs_soma, erro_rel_soma, \
           subtracao, erro_abs_subtracao, erro_rel_subtracao, \
           produto, erro_abs_produto, erro_rel_produto, \
           divisao, erro_abs_divisao, erro_rel_divisao

# Entrada de dados
x_barra = float(input("Digite o valor de x: "))
ex = float(input("Digite a margem de erro de x: "))
y_barra = float(input("Digite o valor de y: "))
ey = float(input("Digite a margem de erro de y: "))

resultado = calcular_operacoes(x_barra, ex, y_barra, ey)

print("\n")
print("Resultado da soma:", resultado[0])
print("Erro absoluto da soma:", resultado[1])
print("Erro relativo da soma:", resultado[2])

print("\n")
print("Resultado da subtração:", resultado[3])
print("Erro absoluto da subtração:", resultado[4])
print("Erro relativo da subtração:", resultado[5])

print("\n")
print("Resultado do produto:", resultado[6])
print("Erro absoluto do produto:", resultado[7])
print("Erro relativo do produto:", resultado[8])

print("\n")
print("Resultado da divisão:", resultado[9])
print("Erro absoluto da divisão:", resultado[10])
print("Erro relativo da divisão:", resultado[11])
```

#Questão 2


```python
import random

def gerar_numeros(n):
    numeros = []
    for i in range(n):
        valor = random.uniform(0, 100)  # Range entre 0 e 100
        erro_abs = valor / 100  # 1% de erro absoluto
        numeros.append((valor, erro_abs))
    return numeros

def calcular_soma_produto(numeros):
    soma = sum(valor for valor, _ in numeros)
    produto = 1
    for valor, _ in numeros:
        produto *= valor

    return soma, produto

def calcular_erros(numeros, soma, produto):
    # Calcular os erros absolutos
    erro_abs_soma = sum(erro for _, erro in numeros)
    erro_abs_produto = sum(erro * abs(produto) for _, erro in numeros)

    # Calcular os erros relativos
    erro_rel_soma = erro_abs_soma / abs(soma)
    erro_rel_produto = erro_abs_produto / abs(produto)

    return erro_abs_soma, erro_rel_soma, erro_abs_produto, erro_rel_produto

n = int(input("Digite o número de valores a serem gerados (n >= 2): "))
print("\n")
if n < 2:
    print("Por favor, insira um número inteiro positivo maior ou igual a 2.")
else:
    numeros = gerar_numeros(n)

    soma, produto = calcular_soma_produto(numeros)

    erro_abs_soma, erro_rel_soma, erro_abs_produto, erro_rel_produto = calcular_erros(numeros, soma, produto)

    print("Números gerados:")
    for i, (valor, erro) in enumerate(numeros, 1):
        print(f"x{i} =", valor, "±", erro)

    print("\nSoma dos números:", soma)
    print("Erro absoluto da soma:", erro_abs_soma)
    print("Erro relativo da soma:", erro_rel_soma)

    print("\nProduto dos números:", produto)
    print("Erro absoluto do produto:", erro_abs_produto)
    print("Erro relativo do produto:", erro_rel_produto)
```

#Questão 3

```python
import random
import matplotlib.pyplot as plt

def gerar_numeros(n):
    numeros = []
    for i in range(n):
        valor = random.randint(1, 100)  # Range entre 0 e 100
        erro_abs = valor / 100  # 1% de erro absoluto
        numeros.append((valor, erro_abs))
    return numeros

def calcular_soma_produto(numeros):
    soma = sum(valor for valor, _ in numeros)
    produto = 1
    for valor, _ in numeros:
        produto *= valor

    return soma, produto

def calcular_erros(numeros, soma, produto):
    # Calcular os erros absolutos
    erro_abs_soma = sum(erro for _, erro in numeros)
    erro_abs_produto = sum(erro * abs(produto) for _, erro in numeros)

    # Calcular os erros relativos
    erro_rel_soma = erro_abs_soma / abs(soma)
    erro_rel_produto = erro_abs_produto / abs(produto)

    return erro_rel_soma, erro_rel_produto

n_values = list(range(2, 21))
es_rel_values = []
ep_rel_values = []

# Gerar numeros e calcular erros relativos
for n in n_values:
    numeros = gerar_numeros(n)
    soma, produto = calcular_soma_produto(numeros)
    es_rel, ep_rel = calcular_erros(numeros, soma, produto)
    es_rel_values.append(es_rel)
    ep_rel_values.append(ep_rel)

# Plotar os gráficos
plt.figure(figsize=(10, 5))

# Gráfico n × ES_REL
plt.subplot(1, 2, 1)
plt.plot(n_values, es_rel_values, marker='o', color='blue')
plt.title('n × ES_REL')
plt.xlabel('n')
plt.ylabel('ES_REL')

# Gráfico n × EP_REL
plt.subplot(1, 2, 2)
plt.plot(n_values, ep_rel_values, marker='o', color='green')
plt.title('n × EP_REL')
plt.xlabel('n')
plt.ylabel('EP_REL')

plt.tight_layout()
plt.show()
```

#Questão 4


```python
import random
import matplotlib.pyplot as plt

def calcular_operacoes(numeros):
    soma = sum(n for n, _ in numeros)
    produto = 1
    for n, _ in numeros:
        produto *= n

    return soma, produto

def calcular_erros(numeros, soma, produto):
    # Calcular os erros absolutos
    erro_abs_soma = sum(erro for _, erro in numeros)
    erro_abs_produto = sum(erro * abs(produto) for _, erro in numeros)

    # Calcular os erros relativos
    erro_rel_soma = erro_abs_soma / abs(soma)
    erro_rel_produto = erro_abs_produto / abs(produto)

    return erro_abs_soma, erro_rel_soma, erro_abs_produto, erro_rel_produto

# Geração de números aleatórios com erro absoluto de 0.1%
def gerar_numeros(n):
    numeros = []
    for i in range(n):
        valor = random.uniform(0, 100)  # Range entre 0 e 100
        erro_abs = valor * 0.001  # 0.1% de erro absoluto
        numeros.append((valor, erro_abs))
    return numeros

n_values = list(range(2, 21))
es_rel_values = []
ep_rel_values = []

for n in n_values:
    numeros = gerar_numeros(n)
    soma, produto = calcular_operacoes(numeros)
    erro_abs_soma, erro_rel_soma, erro_abs_produto, erro_rel_produto = calcular_erros(numeros, soma, produto)
    es_rel_values.append(erro_rel_soma)
    ep_rel_values.append(erro_rel_produto)

# Plotar os gráficos
plt.figure(figsize=(10, 5))

# Gráfico n × ES_REL
plt.subplot(1, 2, 1)
plt.plot(n_values, es_rel_values, marker='o', color='blue')
plt.title('n × ES_REL (Erro Relativo da Soma)')
plt.xlabel('n')
plt.ylabel('ES_REL')

# Gráfico n × EP_REL
plt.subplot(1, 2, 2)
plt.plot(n_values, ep_rel_values, marker='o', color='green')
plt.title('n × EP_REL (Erro Relativo do Produto)')
plt.xlabel('n')
plt.ylabel('EP_REL')

plt.tight_layout()
plt.show()
```
