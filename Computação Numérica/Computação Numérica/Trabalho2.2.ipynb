2) Em uma lista, para cada valor de t da discretização salve o erro absoluto E_abs = | y(t_i) - y_i |

```python
import math

def f(t, y0):
    y = -0.1 * y0
    return y

# definição das variáveis
t0 = 0
y0 = 1.0
h = round(random.random(), 2) # h aleatório entre 0 e 1 com duas casas decimais
n = 5

# listas para armazenar os valores de t e y e os erros absolutos
t_values = [t0]
y_values = [y0]
error_values = [0]

t = t0
y = y0

for i in range(n):
    t = t0 + i * h
    y = y + h * f(t, y)

    y_exact = math.exp(t) #valor exato de y(t)

    error = abs(y_exact - y) #erro absoluto

    #a diciona os valores encontrados as listas
    t_values.append(t)
    y_values.append(y)
    error_values.append(error)

    print("t:", t, "y:", y, "Erro Absoluto:", error)

print("\n")
print("Valores de t:", t_values)
print("Valores de y:", y_values)
print("Erros Absolutos:", error_values)
```

3) Em uma lista, para cada valor de t da discretização salve o erro relativo E_rel =  E_abs / | y(t_i) |

```python
def f(t, y0):
    y = -0.1 * y0
    return y

# definição das variáveis
t0 = 0
y0 = 1.0
h = round(random.random(), 2) # h aleatório entre 0 e 1 com duas casas decimais
n = 5

# listas para armazenar os valores de t, y e os erros absolutos e os erros relativos
t_values = [t0]
y_values = [y0]
error_values = [0]
relative_error_values = [0]

t = t0
y = y0

for i in range(n):
    t = t0 + i * h
    y = y + h * f(t, y)

    y_exact = math.exp(t) #valor exato de y(t)

    error = abs(y_exact - y) #erro absoluto

    relative_error = error / abs(y_exact) #erro relativo

    #adiciona os valores encontrados as listas
    t_values.append(t)
    y_values.append(y)
    error_values.append(error)
    relative_error_values.append(relative_error)

    print("t:", t, "y:", y, "Erro Absoluto:", error, "Erro Relativo:", relative_error)

print("\n")
print("Valores de t:", t_values)
print("Valores de y:", y_values)
print("Erros Absolutos:", error_values)
print("Erros Relativos:", relative_error_values)
```

7) Suponhamos que uma dada população pode ser dividida em duas partes: os que têm uma determinada doença e podem infectar outros, e os que não tem,mas são suscetíveis a tal doença. Seja x a proporção de indivíduos suscetíveis e y a proporção de infectados; logo, x + y = 1. Suponha que a doença se espalhe através do contato entre os indivíduos doentes e os que não estão, e
que a taxa de disseminação dy /dt é proporcional ao número do contato entre infectados e não infectados. Suponha também que os elementos dos dois grupos circulam livremente, de forma que o número de contatos seja proporcional ao produto de x e y. Como temos x = 1 − y, o P.V.I. que modela o problema é dado por

y'(t) = alpha y(t) ( 1- y(t) )
y(0) = y0


Em uma cidade com 10000 habitantes, 10 habitantes estão doentes em um determinado dia (y0 =10). Utilize o Método de Euler para resolver o PVI para diferentes valores de alpha (considere 0 < alpha < 1, alpha = 1 e alpha > 1 )

```python
def f(alpha, y):
    return alpha * y * (1 - y)

def euler(alpha, y0, h, n):
    t_values = [0]
    y_values = [y0]

    t = 0
    y = y0

    for i in range(n):
        t += h
        y += h * f(alpha, y)

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values


y0 = 10  # numero inicial de individuos doentes
h = 0.1  # passo de integração
n = 100  # numero de iterações

# lista de valores de alpha para testar
alphas = [0.1, 0.5, 0.9, 1, 1.1]

# para cada valor de alpha, resolvemos o PVI usando o Método de Euler
for alpha in alphas:
    t_values, y_values = euler(alpha, y0, h, n)

    # plotagem
    plt.plot(t_values, y_values, label=f'alpha={alpha}')

# configurações do gráfico
plt.xlabel('Tempo')
plt.ylabel('Proporção de infectados')
plt.title('Propagação da doença com diferentes valores de alpha')
plt.legend()
plt.grid(True)
plt.show()
```

8) É possível mostrar que a solução do PVI acima é dada por y(t) = y0 / [ (1-y0) exp(-alpha t) + y0 ]. Gere os gráficos definidos nos itens 4, 5 e 6 para o problema do item 7.

```python
# função para calcular a solução analítica
def analytical_solution(alpha, y0, t):
    return y0 / ((1 - y0) * math.exp(-alpha * t) + y0)

# função para calcular o erro absoluto
def absolute_error(estimated, exact):
    return abs(estimated - exact)

# função para calcular o erro relativo
def relative_error(estimated, exact):
    return absolute_error(estimated, exact) / abs(exact)

# função para calcular a taxa de variação
def f(alpha, y):
    return alpha * y * (1 - y)

# função para resolver o PVI usando o Método de Euler
def euler(alpha, y0, h, n):
    t_values = [0]
    y_values = [y0]

    t = 0
    y = y0

    for i in range(n):
        t += h
        y += h * f(alpha, y)

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# definição das variaveis
y0 = 10  # numero inicial de individuos doentes
h = 0.1  # passo de integração
n = 100  # numero de iterações
t_values = [i * h for i in range(n+1)]  # valores de t para o grafico

# lista de valores de alpha para testar
alphas = [0.1, 0.5, 0.9, 1, 1.1]

# para cada valor de alpha, resolvemos o PVI usando o Método de Euler
for alpha in alphas:

    y_analytical = [analytical_solution(alpha, y0, t) for t in t_values] #solução analítica

    t_numerical, y_numerical = euler(alpha, y0, h, n) #solução numérica usando o Método de Euler

    absolute_errors = [absolute_error(y_numerical[i], y_analytical[i]) for i in range(len(t_numerical))] #erros absolutos
    relative_errors = [relative_error(y_numerical[i], y_analytical[i]) for i in range(len(t_numerical))] #erros relativos

    # graficos
    plt.figure(figsize=(15, 5))

    # graf t versus y(t) analítico e numérico
    plt.subplot(1, 3, 1)
    plt.plot(t_values, y_analytical, label='Solução Analítica')
    plt.plot(t_numerical, y_numerical, label='Solução Numérica', linestyle='dashed')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title(f'Solução Analítica vs. Solução Numérica (alpha={alpha})')
    plt.legend()
    plt.grid(True)

    # graf t versus Erro Absoluto
    plt.subplot(1, 3, 2)
    plt.plot(t_numerical, absolute_errors, label='Erro Absoluto')
    plt.xlabel('t')
    plt.ylabel('Erro Absoluto')
    plt.title(f'Erro Absoluto (alpha={alpha})')
    plt.legend()
    plt.grid(True)

    # graf t versus Erro Relativo
    plt.subplot(1, 3, 3)
    plt.plot(t_numerical, relative_errors, label='Erro Relativo')
    plt.xlabel('t')
    plt.ylabel('Erro Relativo')
    plt.title(f'Erro Relativo (alpha={alpha})')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
